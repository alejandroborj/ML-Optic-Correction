```python
from cpymad.madx import Madx
import tfs
import matplotlib.pyplot as plt
```

```python
madx = Madx()

madx.call(file='job.create_model.madx')
```

```python
#myDF_LHC=madx.table.twiss.dframe()

df = tfs.reader.read_tfs("twiss.dat")
```

```python
# plotting the results
plt.plot(df.S,df.BETX,'o-b', label='$\\beta_x$')
plt.plot(df.S,df.BETY,'o-r', label='$\\beta_y$')
plt.ylim(0,300)
plt.xlabel('s [m]')
plt.ylabel('[m]')
plt.grid()
plt.legend()
```
